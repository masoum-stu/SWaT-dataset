```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

file_path = '/content/drive/MyDrive/D2_corrected-1.csv'
data = pd.read_csv(file_path)

# Convert 'Timestamp' to datetime
data[' Timestamp'] = pd.to_datetime(data[' Timestamp'])

# Set the Timestamp as the index
data.set_index(' Timestamp', inplace=True)

# Assuming the time range and intervals are already in the correct format
# We'll filter the data to the time range we're interested in
filtered_data = data.between_time('11:21:00', '11:29:00')

# Plotting the process behavior of LIT101
plt.figure(figsize=(10, 4))
plt.plot(filtered_data.index, filtered_data['LIT101'], label='LIT101', color='green', marker='o')

# Set the x-axis major locator to 2-minute intervals and formatter
plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=2))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

# Rotate and align the tick labels so they look better
plt.gcf().autofmt_xdate()

plt.xlabel('Time')
plt.ylabel('LIT101 Level')
plt.title('Process Behavior of LIT101')
plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```
